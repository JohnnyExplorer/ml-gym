```python
clarity_dict = {'IF' : 7,
'VVS1' :6 ,
'VVS2' :5 , 
'VS1' : 4,
'VS2' : 3.,
'SI1' : 2,
'SI2' : 1,
'I1' :0}

data['clarity'].replace(clarity_dict, inplace=True)

data.head()
```

```python
numeric_features = ['carat','depth','table','x','y','z']
numeric_df = data[numeric_features]
numeric_df.describe()
```

```python
from sklearn.preprocessing import scale
numeric_array = scale(numeric_df)
numeric_array[:5]
```

```python
numeric_df = pd.DataFrame(numeric_array, columns=numeric_df.columns)
numeric_df.describe()
```

```python
data.drop(numeric_features, inplace=True, axis=1)
data.head()
```

```python
data = pd.concat([data,numeric_df], axis=1)
data.shape
```

```python
data.head()
```

```python
data.to_csv('datasets/diamonds_pro.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
X = data.drop('price',axis=1)
Y = data['price']

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
linear_regression.fit(x_train,y_train)
```

```python
y_prediction = linear_regression.predict(x_test)
y_prediction
```

```python
train_score = linear_regression.score(x_train,y_train)
print("train score : ", train_score)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_prediction)
print("test score : ", score)
```
